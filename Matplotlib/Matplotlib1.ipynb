### 2D LINE PLOT 
>.Bivariate Analysis

>.Categorical->numerical and numerical->categorical

>. Use Case-Time Series Data

```python
#plotting a simple function
price=[48000,54000,57000,49000,47000,45000]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
#from a pandas dataframe
batsman=pd.read_csv('sharma-kohli.csv')
batsman.head()
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#labeling the axes
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#colour and style
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#line styles
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2)
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3)
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#markers
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#legend and location
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d',label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o',label='Kohli')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
```

```python
#limiting the x and y axis
price=[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,75000)
```

```python
#grids 
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d',label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o',label='Kohli')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
```

```python
#show
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d',label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o',label='Kohli')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

# Bar Chart 
>.Numerical Vs Categorical

>.Use Case-Aggregate Analysis Of Groups

```python
# simple bar chart
children=[10,20,30,40,50]
colors=['red','green','blue','yellow','pink']
plt.bar(colors,children)
```

```python
#horizontal bar chart
children=[10,20,30,40,50]
colors=['red','green','blue','yellow','pink']
plt.barh(colors,children)
```

```python
# color and label
df=pd.read_csv('batsman_season_record.csv')
df.head()
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#problem scenario
children=[10,20,30,40,50]
colors=['red red red red red','green green green green','blue blue blue blue','yellow yellow yellow yellow','pink pink pink pink']
plt.bar(colors,children)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#stacked bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')
plt.legend()
plt.show()
```

# PIE CHART

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
#simple data
d=[23,45,67,89,98,12,34]
subjects=['eng','maths','physics','chem','bio','hindi','sst']
plt.pie(d,labels=subjects)
plt.show()
```

```python
df=pd.read_csv('gayle-175.csv')
df.head()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','green','blue','pink','yellow','pink','blue'])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','green','blue','pink','yellow','pink','blue'],explode=[0,0.1,0,0,0,0])
plt.show()
```
