### 2D LINE PLOT 
>.Bivariate Analysis

>.Categorical->numerical and numerical->categorical

>. Use Case-Time Series Data

```python
#plotting a simple function
price=[48000,54000,57000,49000,47000,45000]
year=[2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
#from a pandas dataframe
batsman=pd.read_csv('sharma-kohli.csv')
batsman.head()
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#labeling the axes
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#colour and style
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#line styles
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2)
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3)
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#markers
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
#legend and location
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d',label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o',label='Kohli')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
```

```python
#limiting the x and y axis
price=[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,75000)
```

```python
#grids 
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d',label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o',label='Kohli')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
```

```python
#show
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='--',linewidth=2,marker='d',label='Rohit')
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='--',linewidth=3,marker='o',label='Kohli')
plt.title('Batsman Performance')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
plt.grid()
plt.show()
```

## Scatter PLots
.Use Case-Finding Correlation

```python
#plt.scatter simple function
x=np.linspace(-10,10,50)
y=10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
#scatter from a pandas dataframe
batter=pd.read_csv('batter.csv')
df=batter.head(50)
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='s')
plt.title('Avg and Strike Rate')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
```

```python
#size
tips=sns.load_dataset('tips')
tips.head()
plt.scatter(tips['total_bill'],tips['tip'],color='red',marker='o',s=tips['size']*20)
```
