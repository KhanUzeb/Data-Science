```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## COLORED SCATTERPLOTs

```python
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="winter",alpha=0.8)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.figure(figsize=(14,7))
plt.scatter(iris["SepalLengthCm"],iris["PetalLengthCm"],c=iris["Species"],cmap="winter",alpha=0.8)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
```

```python
batters=pd.read_csv("batter.csv")
batters.head()
sample_df=batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## HORIZONTAL AND VERTICAL LINE

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='r')
plt.axvline(30,color='b')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### SUBPLOTS

```python
batters.head()
```

```python
plt.scatter(batters['avg'],batters['strike_rate'])
```

```python
fig,ax=plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title("Scatter plot of average and strike rate")
ax.set_xlabel("Average")
ax.set_ylabel("Strike rate")
fig.show()
```

```python
#subplots
fig,ax=plt.subplots(2,1,sharex=True,figsize=(15,6))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='r')
ax[1].scatter(batters['avg'],batters['runs'])
ax[0].set_title("Scatter plot of average and strike rate")
ax[0].set_xlabel("Average")
ax[0].set_ylabel("Strike rate")
ax[1].set_title("Scatter plot of average and runs")
ax[1].set_xlabel("Average")
ax[1].set_ylabel("Runs")
fig.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='r')
ax[1,0].scatter(batters['avg'],batters['runs'])
ax[0,0].set_title("Scatter plot of average and strike rate")
ax[0,0].set_xlabel("Average")
ax[0,0].set_ylabel("Strike rate")
ax[1,0].set_title("Scatter plot of average and runs")
ax[1,0].set_xlabel("Average")
ax[1,0].set_ylabel("Runs")
ax[0,1].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
fig.show()
```
