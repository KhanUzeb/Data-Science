## Group By 

```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('imdb-top-1000.csv')
movies.head()
```

```python
genres=movies.groupby('Genre')
```

```python
#apply built-in functions
genres.sum()
```

```python
#find the top 3 genres by total earnings
movies.groupby('Genre')['Gross'].sum().nlargest(3)
```

```python
#find the genre with the highest average rating
movies.groupby('Genre')['IMDB_Rating'].mean().nlargest(1)
```

```python
#find the director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().nlargest(1)
```

```python
#find highest rated movie in each genre
highest_rated_per_genre = movies.loc[movies.groupby('Genre')['IMDB_Rating'].idxmax()]
highest_rated_per_genre[['Genre', 'Series_Title', 'IMDB_Rating']].reset_index(drop=True)
```

```python
#find movies done by each star
movies.groupby('Star1')['Series_Title'].count().nlargest(5)
```

```python
#agg methods
genres.agg({
    'Runtime':['mean', 'max', 'min'],
    'IMDB_Rating':['mean', 'max', 'min'],
    'Gross':['sum', 'mean', 'max', 'min'],
    'No_of_Votes':['sum', 'mean', 'max', 'min']
})
```

```python
#split-apply-combine
genres.apply(min)
```

```python
#find number of movies which start with a for each genre
def f(group):
    return group['Series_Title'].str.startswith('A').sum()
genres.apply(f)
```

```python
#find ranking of each movie in its genre based on IMDB rating
def rank_movie(group):
    group['Genre_Rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
genres.apply(rank_movie)
```

```python
#find the normalized IMDB rating for each movie in its genre
def normal(group):
    group['Normalized_Rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].mean()) / group['IMDB_Rating'].std()
    return group
genres.apply(normal)
```

```python
#groupby with multiple columns
duo=movies.groupby(['Director','Star1'])
duo.size()
duo.get_group(('Christopher Nolan', 'Christian Bale'))
```

```python
#find the most earning duo
duo['Gross'].sum().nlargest(1)
```

```python
#find the best duo in terms of metascore actor-genre combo
duo=movies.groupby(['Star1','Genre'])
duo['Metascore'].mean().reset_index().sort_values(by='Metascore', ascending=False).head(10)
```

```python
ipl=pd.read_csv('deliveries.csv')
ipl.head()
ipl.shape
```

```python
#find the top 10 players by total runs
top_players = ipl.groupby('batsman')['batsman_runs'].sum().nlargest(10)
top_players
```

```python
#find the batman with max no of 6s
six=ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman_runs'].count().nlargest(1).index[0]
```

```python
#find the most number of 4s and 6s by a player in last 5 overs
temp = ipl[ipl['over'] >= 16]
temp[(temp['batsman_runs'] == 4) | (temp['batsman_runs']==6)]
temp.groupby('batsman')['batsman_runs'].count().nlargest(5)
```

```python
#find v kohli record againt all teams
kohli = ipl[ipl['batsman'] == 'V Kohli']
kohli.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
#create a function that can return the highest run score of any batsman
temp=ipl[ipl['batsman']== 'V Kohli']
temp.groupby('match_id')['batsman_runs'].sum().sort_values (ascending=False).head(1).values[0]
```

```python
def find_highest_run_score(batsman_name):
    temp = ipl[ipl['batsman'] == batsman_name]
    return temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
find_highest_run_score('V Kohli')
```
