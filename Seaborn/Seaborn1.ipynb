### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
```

```python
#scatter plot axis level function
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
plt.show()
```

```python
#relplot is a fig level function -> Square Shape
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',hue='sex',style='time',size='size')
```

```python
#line plot
gap=px.data.gapminder()
gap.head()
```

```python
df=gap[gap['country']=='India']
```

```python
sns.lineplot(x='year',y='lifeExp',data=df)
```

```python
#fig level function
sns.relplot(x='year',y='lifeExp',data=df,kind='line')
```

```python
#hue->style
rdf=gap[gap['country'].isin(['India','Brazil','Germany'])]
```

```python
sns.relplot(x='year',y='lifeExp',data=rdf,kind='line',hue='country',style='continent')
```

```python
#facet plot works only for fig level function
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',col='sex',style='time',size='size',row='smoker')
```

```python
#col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

- figure ->displot

- axes level ->histplot ->kdeplot ->rugplot

```python
#plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
#bins parameter
sns.displot(data=tips,x='total_bill',kind='hist',bins=20)
```

```python
#kde plot
sns.displot(data=tips,x='day',kind='hist')
```

```python
#hue
sns.displot(data=tips,x='tip',kind='hist',hue='sex')
```

```python
#element
sns.displot(data=tips,x='tip',kind='hist',element='step',hue='sex')
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='sex')
```

```python
#facet plot
sns.displot(data=titanic,x='age',kind='hist',hue='sex',col='class')
```

```python
#kdeplot -> density plot
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python
#rugplot
sns.rugplot(data=tips,x='total_bill')
sns.kdeplot(data=tips,x='total_bill')
```

```python
#bivariate histogram
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
#bivariate kdeplot
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='autumn_r')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris.head()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]],cmap='winter')
```

### Figure level function -> `catplot`

```python
#strip plot
#axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
#using catplot
#fig level function
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
#jitter
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.2)
```

```python
#swarm plot
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

![alt text](download.png)

```python
#box plot
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
#using catplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',hue='smoker')
```

```python
#single box plot
sns.catplot(data=tips,y='total_bill',kind='box')
```

### Violinplot = (Boxplot + KDEplot)

```python
#vionlin plot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
#hue
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
# barplot
# some issue with errorbar
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
#pointplot
sns.pointplot(data=tips, x='sex', y='total_bill')
```

```python
#countplot
sns.countplot(data=tips, x='sex',hue='day')
```

```python
#faceting using catplot
sns.catplot(data=tips, x='sex',y='total_bill',col='day',kind='box',row='time')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
#regplot
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
#residplot
sns.residplot(data=tips,x='total_bill',y='tip')
```

## FACET GRID AND MULTI GRID

```python
#fig level -> relplot -> displot ->catplot
g=sns.FacetGrid(tips,col='day',row='time',)
g.map(sns.violinplot,'sex','tip')
```

## PAIR PLOT

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
g=sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
#map_diag -> map_offdiag
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
```

```python
g=sns.PairGrid(iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.barplot)
```
